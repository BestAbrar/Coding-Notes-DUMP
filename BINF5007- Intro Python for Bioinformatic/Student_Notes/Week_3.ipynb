```python
#What if you just wanted to get part of a list?
#We want  TP53-CFTR
tested_genes = ["BRCA1", "TP53", "EGFR", "CFTR", "TNF"]
sub_list = tested_genes[1:3] #This is called slicing
#However, we won't get what we want with this, we will be missing the last index we want
#slicing -> my_list[start_index:last_index+1] - you need to add one more in order to get the last index you want
#so you need to go up one more index
sub_list = tested_genes[1:4]
print(sub_list)

#So if we want everything except the first it'll look like this - even tho we technically only have 0-4 in the index
sub_list = tested_genes[1:5]
print(sub_list)

#If slicling list look like [,3] or [1,]
#It will go from beginging to the 4th index and from the first index to the end
```

```python
tested_genes = ["BRCA1", "TP53", "EGFR", "CFTR", "TNF"]
#What if I wanted to BRCA22 in after BRCA1
tested_genes[1:2] = ["BRCA2"] #This won't work as it will replace TP53
print(tested_genes)

#So you need to write it like this - double the index where you want it to go
tested_genes = ["BRCA1", "TP53", "EGFR", "CFTR", "TNF"]
tested_genes[1:1] = ["BRCA2"] #inserts with slicing. also works with [3:3] and [4:4] etc
print(tested_genes)

#This works with multiple variables as well
tested_genes = ["BRCA1", "TP53", "EGFR", "CFTR", "TNF"]
tested_genes[5:5] = ["BRCA2", "AP0E"] 
print(tested_genes)

#If you don't put anything in the [], it will replace the whole list
tested_genes = ["BRCA1", "TP53", "EGFR", "CFTR", "TNF"]
tested_genes[:] = ["BRCA2", "AP0E"] 
print(tested_genes)

#Can make an empty list by just leaving empty square brackets
tested_genes = []
```
